# 2. [Chapter 01] 파이썬 기반의 머신러닝과 생태계 이해

## 2.1 머신러닝의 개념

- 일반적으로는 데이터를 기반으로 패턴을 학습하고 결과를 예측하는 알고리즘 기법을 의미


- 지도학습 : 분류, 회귀, 추천 시스템, 시각/음성 감지/인지, 텍스트 분석


- 비지도학습 : 클러스터링, 차원 축소, 강화 학습

## 2.2 파이썬 머신러닝 패키지

- 머신러닝 패키지, 행렬/선형대수/통계 패키지, 데이터 핸들링, 시각화

## 2.3 넘파이

- array() : 다양한 인자를 입력 받아 ndarray로 변환하는 기능 수행

    
- arange() : 0부터 함수 인자 값 -1까지의 값을 ndarray의 데이터값으로 변환


- zeros() : 모든 값을 0으로 채운 ndarray 반환
    
    
- ones() : 모든 값을 1로 채운 ndarray 반환
    
    
- reshape() : ndarray를 특정 차원 및 크기로 변환


- 인덱싱 : 특정 데이터만 추출, 슬라이싱, 팬시 인덱싱, 불린 인덱싱


- sort() : 넘파이에서 행렬을 정렬하는 대표적인 방법


- argsort() : 정렬된 행렬의 인덱스 반환


- dot() : 행렬의 내적 계산


- transpose() : 전치 행렬 반환

## 2.4 데이터 핸들링 – 판다스

- head() : DataFrame의 맨 앞에 있는 N개의 로우 반환


- info() : 총 데이터 건수, 데이터 타입, Null 건수 반환


- describe() : 숫자형 칼럼의 n-percentile 분포도, 평균값, 최댓값, 최솟값 반환


- value_counts() : 칼럼값의 유형과 건수 (데이터의 분포도) 확인


- tolist(), to_dict() : 각각 리스트, 딕셔너리로 변환


- drop() : 데이터 삭제 (axis 0: 로우 방향, axis 1: 칼럼 방향)


- reset_index() : 새롭게 인덱스를 연속 숫자형으로 할당


- iloc() : 위치 기반 인덱싱, 정수형 값 지정


- loc() : 명칭 기반 인덱싱, 시작점에서 종료점을 포함한 데이터 반환


- sort_values() : DataFrame, Series 정렬


- Aggregation 함수 : min(), max(), sum(), count()


- groupby() : 해당 칼럼을 기준으로 groupby 된 객체 반환


- isna() : NaN 여부 확인


- fillna() : NaN 값을 다른 값으로 대체
