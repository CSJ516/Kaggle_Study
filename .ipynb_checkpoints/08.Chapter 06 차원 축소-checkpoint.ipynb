# 1. [Chapter 06] 차원 축소

## 1.1 차원 축소 개요

- 차원 축소 : 매우 많은 피처로 구성된 다차원 데이터 세트의 차원을 축소해 새로운 차원의 데이터 세트를 생성하는 것

피처가 많을 경우 개별 피처 간에 상관관계가 높을 가능성이 크며, 선형 모델에서는 이로 인한 다중 공선성 문제가 발생할 수 있음

다차원의 피처를 차원 축소해 피처 수를 줄이면 **더 직관적으로 데이터를 해석할 수 있으며, 학습 데이터의 크기를 줄여 높은 처리 능력이 요구되지 않음**

    피처 선택 : 특정 피처에 종속성이 강한 불필요한 피처를 제거하고, 데이터의 특징을 잘 나타내는 주요 피처만 선택

    피처 추출 : 기존 피처를 저차원의 중요 피처로 압축해서 추출 (기존의 피처와는 완전히 다른 값)
    
**단순한 데이터의 압축이 아닌 잠재적인 요소를 추출하는 것**으로, 이미지나 텍스트에서 주로 활용 (ex) PCA, SVD, NMF)

## 1.2 PCA

- **PCA(주성분 분석) : 대표적인 차원 축소 기법. 여러 변수 간의 상관관계를 이용해 이를 대표하는 주성분을 추출하여 차원 축소**

분산이 데이터의 특성을 가장 잘 나타내는 것으로 간주하고, 데이터 변동성이 가장 큰 방향으로 축을 생성하여 데이터를 투영하는 방식

    1) 입력 데이터 세트의 공분산 행렬 생성
    
    2) 공분산 행렬의 고유벡터와 고유값 계산
    
    3) 고유값이 가장 큰 순으로 K개(PCA 변환 차수)만큼 고유벡터를 추출
    
    4) 3)에서 추출된 고유벡터를 이용해 새롭게 입력 데이터를 선형 변환

```python
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

iris = load_iris()
columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
df = pd.DataFrame(iris.data, columns=columns)
df['target'] = iris.target
df.head(3)
```

```python
markers = ['^','s','o']
# setosa = 0, versicolor = 1, virginica = 2
for i, marker in enumerate(markers):
    x_axis = df[df['target'] == i]['sepal_length']
    y_axis = df[df['target'] == i]['sepal_width']
    plt.scatter(x_axis, y_axis, marker=marker, label=iris.target_names[i])
plt.legend()
plt.xlabel('sepal length')
plt.ylabel('sepal_width')
plt.show()
```

PCA는 속성의 스케일에 영향을 받으므로, **StandardScaler**를 이용해 각 속성값을 동일한 표준 정규 분포로 변환

```python
from sklearn.preprocessing import StandardScaler

scaled_df = StandardScaler().fit_transform(df.iloc[:, :-1])     # Target 값 제외
```

**PCA 클래스**로 4차원(4개 속성)의 붓꽃 데이터를 2차원(2개의 PCA 속성) 데이터로 변환

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
iris_pca = pca.fit_transform(scaled_df)
print(iris_pca.shape)
```

```python
pca_columns = ['pca_1', 'pca_2']
pca_df = pd.DataFrame(iris_pca, columns=pca_columns)
pca_df['target'] = iris.target
pca_df.head(3)
```

2개의 속성으로 PCA 변환된 데이터 세트를 2차원상에서 시각화

```python
markers = ['^','s','o']
for i, marker in enumerate(markers):
    x_axis = pca_df[pca_df['target'] == i]['pca_1']
    y_axis = pca_df[pca_df['target']== i]['pca_2']
    plt.scatter(x_axis, y_axis, marker=marker, label=iris.target_names[i])
plt.legend()
plt.xlabel('pca_1')
plt.ylabel('pca_2')
plt.show()
```

PCA로 변환하기 전에는 versicolor와 virginica 품종이 잘 구분되지 않았으나, 변환 후 비교적 잘 구분된 모습을 볼 수 있음

pca_1이 원본 데이터의 변동성을 잘 반영했기 때문이며, **explained_variance_ratio_** 속성으로 개별 PCA component별로 차지하는 변동성 비율 파악

```python
print(pca.explained_variance_ratio_)
```

PCA를 2개 요소로만 변환해도 원본 데이터의 변동성을 95% 설명할 수 있음

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
import numpy as np

# 원본 붓꽂 데이터 세트에 분류 적용
rcf = RandomForestClassifier(random_state=156)
scores = cross_val_score(rcf, iris.data, iris.target, scoring='accuracy', cv=3)
print('원본 데이터 교차 검증 개별 정확도: ', scores)
print('원본 데이터 평균 정확도: ', np.mean(scores))

# PCA로 변환된 데이터 세트에 분류 적용
pca_X = pca_df[['pca_1', 'pca_2']]
pca_scores = cross_val_score(rcf, pca_X, iris.target, scoring='accuracy', cv=3)
print('PCA 변환 데이터 교차 검증 개별 정확도: ', pca_scores)
print('PCA 변환 데이터 평균 정확도: ', np.mean(pca_scores))
```

정확도가 10% 정도 하락하였으나, 속성 개수가 반으로 감소한 것을 고려한다면 원본 데이터의 특성을 상당 부분 유지하고 있음을 알 수 있음

### 1.2.1 UCI 신용카드 고객 데이터 세트에 PCA 적용하기

(https://archive.ics.uci.edu/ml/datasets/default+of+credit+card+clients)

```python
import pandas as pd

df = pd.read_excel('./data/default of credit card clients.xls', header=1, sheet_name='Data').iloc[0:,1:]
print(df.shape)
df.head(3)
```

default payment next month(다음달 연체 여부)가 Target 값이며, 연체일 경우 1 정상납부일 경우 0

PAY_0 -> PAY_1, default payment next month -> default 로 칼럼명 변경

```python
df.rename(columns={'PAY_0': 'PAY_1', 'default payment next month': 'default'}, inplace=True)
y_target = df['default']
X_features = df.drop('default', axis=1)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

corr = X_features.corr()
plt.figure(figsize=(14, 14))
sns.heatmap(corr, annot=True, fmt='.1g')     # fmt로 소수점 조절
```

상관도가 0.8 ~ 0.9 로 높은 BILL_AMT 6개 속성을 2개의 component로 PCA 변환

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

cols_bill = ['BILL_AMT'+str(i) for i in range(1,7)]
print('대상 속성명 : ', cols_bill)

scaler = StandardScaler()
scaled_df = scaler.fit_transform(X_features[cols_bill])
pca = PCA(n_components=2)
pca.fit(scaled_df)
print('PCA Component별 변동성: ', pca.explained_variance_ratio_)
```

2개의 PCA Component만으로도 6개 속성의 변동성을 약 95% 이상 설명

```python
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

# 원본 데이터 세트에 분류 적용
rcf = RandomForestClassifier(n_estimators=300, random_state=156)
scores = cross_val_score(rcf, X_features, y_target, scoring='accuracy', cv=3)
print('원본 데이터 교차 검증 개별 정확도: ', scores)
print('원본 데이터 평균 정확도: {:.4f}'.format(np.mean(scores)))

# PCA로 변환된 데이터 세트에 분류 적용
scaler = StandardScaler()
scaled_df = scaler.fit_transform(X_features)
pca = PCA(n_components=6)
pca_df = pca.fit_transform(scaled_df)
pca_scores = cross_val_score(rcf, pca_df, y_target, scoring='accuracy', cv=3)
print('PCA 변환 데이터 교차 검증 개별 정확도: ', pca_scores)
print('PCA 변환 데이터 평균 정확도: {:.4f}'.format(np.mean(pca_scores)))
```

전체 23개 속성의 1/4 수준인 6개의 component만으로도 원본 데이터를 기반으로 한 결과보다 약 2%의 성능 저하 발생

PCA는 컴퓨터 비전, 특히 얼굴 인식의 경우 Eigen-face라고 불리는 PCA 변환으로 많이 사용됨

## 1.3 LDA

- **LDA(선형 판별 분석) : PCA와 매우 유사하며, 개별 클래스를 분별할 수 있는 기준을 최대한 유지하면서 차원 축소**

PCA는 입력 데이터의 변동성을 가장 큰 축으로 찾지만, LDA는 입력 데이터의 결정 값 클래스를 최대한으로 분리할 수 있는 축을 찾음

클래스 간 분산은 최대한 크게, 클래스 내부의 분산은 최대한 작게 가져가는 방식

    1) 입력 데이터의 결정 값 클래스별로 개별 피처의 평균 벡터를 기반으로 클래스 내부와 클래스 간 분산 행렬을 구함
    
    2) 고유값이 가장 큰 순으로 K개(LDA변환 차수)만큼 고유벡터를 추출
    
    3) 2)에서 추출된 고유벡터를 이용해 새롭게 입력 데이터 변환

### 1.3.1 붓꽃 데이터 세트에 LDA 적용하기

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris

iris = load_iris()
scaled_df = StandardScaler().fit_transform(iris.data)
```

**LDA는 PCA와 다르게 비지도학습이 아닌 지도학습이므로, 변환 시에 클래스의 결정 값이 필요**

```python
lda = LinearDiscriminantAnalysis(n_components=2)
lda.fit(scaled_df, iris.target)
lda_df = lda.transform(scaled_df)
print(lda_df.shape)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

lda_columns = ['lda_1', 'lda_2']
lda_iris_df = pd.DataFrame(lda_df, columns=lda_columns)
lda_iris_df['target'] = iris.target

markers = ['^','s','o']
for i, marker in enumerate(markers):
    x_axis = lda_iris_df[lda_iris_df['target'] == i]['lda_1']
    y_axis = lda_iris_df[lda_iris_df['target'] == i]['lda_2']
    plt.scatter(x_axis, y_axis, marker=marker, label=iris.target_names[i])
plt.legend(loc='upper right')
plt.xlabel('lda_1')
plt.ylabel('lda_2')
plt.show()
```

## 1.4 SVD

- SVD(특이값 분해) : PCA와 유사한 행렬 분해 기법을 이용하지만, 정방행렬뿐만 아니라 행과 열의 크기가 다른 행렬에도 적용 가능


- Truncated SVD : 특이값 중 상위 일부 데이터만 추출해 분해하는 방식. 원본 행렬을 정확하게 다시 원복할 수는 없으나 상당한 수준으로 근사할 수 있음

**SVD와 NMF는 매우 많은 피처 데이터를 가진 고차원 행렬을 두 개의 저차원 행렬로 분리하는 행렬 분해 기법**

PCA와 유사하게 이미지 압축을 통한 패턴 인식과 신호 처리 분야에 사용되며, 텍스트의 토픽 모델링 기법인 LSA의 기반 알고리즘

## 1.5 NMF

- **NMF : Truncated SVD와 같이 낮은 랭크를 통한 행렬 근사 방식의 변형**

원본 행렬 내의 모든 원소 값이 모두 양수(0 이상)라는 게 보장되면 좀 더 간단하게 두 개의 기반 양수 행렬로 분해될 수 있는 기법

SVD와 유사하게 이미지 압축을 통한 패턴 인식, 텍스트의 토픽 모델링 기법, 문서 유사도 및 클러스터링에 활용

cf) 사용자-평가 순위 데이터 세트를 행렬 분해 기법을 통해 분해하면서 사용자가 평가하지 않은 상품에 대한 잠재적 요소를 추출해 이를 통해 평가 순위를 예측하고, 높은 순위로 예측된 상품을 추천해주는 방식에도 활용됨 (잠재 요소 기반의 추천 방식)

```python
from sklearn.decomposition import NMF
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')

iris = load_iris()
iris_ftrs = iris.data
nmf = NMF(n_components=2)
nmf.fit(iris_ftrs)
iris_nmf = nmf.transform(iris_ftrs)
plt.scatter(x=iris_nmf[:,0], y=iris_nmf[:,1], c=iris.target)
plt.xlabel('nmf_1')
plt.ylabel('nmf_2')
```
