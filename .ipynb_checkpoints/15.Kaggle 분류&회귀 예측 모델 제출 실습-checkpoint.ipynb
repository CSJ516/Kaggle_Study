# 1. 분류 실습 - 캐글 신용카드 사기 검출

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('./data/creditcard.csv')
print(df.shape)
df.head(3)
```

```python
# 데이터 유형 및 결측치 확인
df.info()
```

```python
df.describe()
```

```python
# Target 값과 독립변수들 간의 상관관계
plt.rcParams['figure.figsize'] = (15,10)

sns.heatmap(df.corr().abs(), cmap='Greens', vmin=0, vmax=1)
```

로지스틱 회귀에서는 다중 공선성을 확인할 필요가 없으나, 공부 목적으로 실습 진행

```python
# 다중 공선성 확인 (VIF : 분산팽창요인)
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
vif['VIF Factor'] = [variance_inflation_factor(df.values, i) for i in range(df.shape[1])]
vif['features'] = df.columns
vif
```

VIF가 5보다 크면 주의, 10보다 크면 위험으로 제거해야 함

단 유사한 feature의 경우 둘 중 1개만 제거하고, VIF계수를 재출력하여 확인하는 과정을 반복

```python
# VIF 계수 높은 feature 제거
df_corr = df.copy()
df_corr = df_corr.drop(['Amount'], axis=1)
vif = pd.DataFrame()
vif['VIF Factor'] = [variance_inflation_factor(df_corr.values, i) for i in range(df_corr.shape[1])]
vif['features'] = df_corr.columns
vif
```

```python
# 이상치에 덜 민감한 RobustScaler 적용
from sklearn.preprocessing import RobustScaler

X_features = df_corr.drop(['Class'], axis=1)
y_target = df_corr['Class']

robustScaler = RobustScaler()
df_corr_rs = robustScaler.fit_transform(X_features)
```

```python
# Target 값의 분포 확인
y_target.value_counts()
```

분포가 불균형하므로 데이터셋 분리 시 stratify 적용

0,1로 분리할 수 있는 로지스틱 회귀 모델 적용

```python
# 데이터셋 분리 및 로지스틱 회귀 모델 적용
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

X_train, X_test, y_train, y_test = train_test_split(X_features, y_target, test_size=0.2, \
                                                    stratify=y_target, random_state=0)
lr = LogisticRegression()
model = lr.fit(X_train, y_train)
pred = model.predict(X_test)
```

```python
# 모델 평가
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score

print('accuracy_score: ', accuracy_score(y_test, pred))
print('recall_score: ', recall_score(y_test, pred))
print('pricision_score: ', precision_score(y_test, pred))
print('f1_score: ', f1_score(y_test, pred))
```

```python
# 다중 공선성을 제거하지 않은 버전
# 이상치에 덜 민감한 RobustScaler 적용
from sklearn.preprocessing import RobustScaler

X_features = df.drop(['Class'], axis=1)
y_target = df['Class']

robustScaler = RobustScaler()
df_corr_rs = robustScaler.fit_transform(X_features)

# 데이터셋 분리 및 로지스틱 회귀 모델 적용
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

X_train, X_test, y_train, y_test = train_test_split(X_features, y_target, test_size=0.2, \
                                                    stratify=y_target, random_state=0)
lr = LogisticRegression()
model = lr.fit(X_train, y_train)
pred = model.predict(X_test)

# 모델 평가
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score

print('accuracy_score: ', accuracy_score(y_test, pred))
print('recall_score: ', recall_score(y_test, pred))
print('pricision_score: ', precision_score(y_test, pred))
print('f1_score: ', f1_score(y_test, pred))
```

# 2.  회귀 실습 - 캐글 레스토랑 수입 예측

```python
res_train = pd.read_csv('./data/restaurant-revenue-prediction/train.csv')
print(res_train.shape)
res_train.head(3)
```

```python
res_test = pd.read_csv('./data/restaurant-revenue-prediction/test.csv')
print(res_test.shape)
res_test.head(3)
```

```python
res_train.info()
```

```python
# Open Date와 현재 날짜의 차이를 계산하여 칼럼 추가
from datetime import datetime

res_train['Open Date'] = res_train['Open Date'].apply(lambda x: x.replace('/',''))
res_train['Open Date'] = res_train['Open Date'].apply(lambda x: datetime.strptime(x, '%m%d%Y'))

res_train['Date Diff'] = 0
for i in range(len(res_train)):
    res_train['Date Diff'][i] = (datetime.now() - res_train['Open Date'][i]).days
    
res_train.head(3)
```

```python
res_test['Open Date'] = res_test['Open Date'].apply(lambda x: x.replace('/',''))
res_test['Open Date'] = res_test['Open Date'].apply(lambda x: datetime.strptime(x, '%m%d%Y'))

res_test['Date Diff'] = 0
for i in range(len(res_test)):
    res_test['Date Diff'][i] = (datetime.now() - res_test['Open Date'][i]).days
    
res_test.head(3)
```

```python
# Pycaret으로 적합한 모델 탐색
from pycaret.regression import *

data_set = setup(res_train, target='revenue', ignore_features=['Id', 'Open Date'])
```

```python
best_models = compare_models(sort='RMSE', n_select=3)
```

```python
best_models
```

```python
tuned_top3 = [tune_model(i, optimize='RMSE') for i in best_models]
```

```python
blender_top3 = blend_models(estimator_list=tuned_top3)
```

```python
final_model = finalize_model(blender_top3)
pred = predict_model(final_model, data=res_test)
```

```python
# 예측한 값으로 submission 파일 생성
submission = pd.read_csv('./data/restaurant-revenue-prediction/sampleSubmission.csv')
submission['Prediction'] = pred['Label']
submission.head(3)
```

```python
submission.to_csv('./data/restaurant-revenue-prediction/20210504_pycaret.csv', index=False)
```
