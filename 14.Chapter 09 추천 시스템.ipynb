# 1. [Chapter 09] 추천 시스템

## 1.1 추천 시스템의 개요와 배경

- 추천 시스템의 유형 : 콘텐츠 기반 필터링 방식, 협업 필터링 방식(최근접 이웃 협업 필터링, 잠재 요인 협업 필터링)

초창기에는 콘텐츠 기반이나 최근접 이웃 협업 필터링이 주로 사용됐지만, 현재 대부분의 온라인 스토어에서 잠재 요인 협업 필터링 적용

개인화 특성을 좀 더 강화하기 위해 하이브리드 형식으로 콘텐츠 기반과 협업 기반을 적절히 결합하는 경우도 있음

## 1.2 콘텐츠 기반 필터링 추천 시스템

- **콘텐츠 기반 필터링 방식 : 사용자가 선호하는 아이템과 비슷한 콘텐츠를 가진 다른 아이템을 추천하는 방식** ex) 비슷한 장르, 감독, 배우

## 1.3 최근접 이웃 협업 필터링

- 협업 필터링 방식 : 사용자가 평가한 아이템을 기반으로 아직 평가하지 않은 아이템의 예측 평가를 도출하여 추천하는 방식

최근접 이웃 방식과 잠재 요인 방식 모두 사용자-아이템 평점 행렬 데이터에만 의지해 추천을 수행

사용자-아이템 평점 행렬의 행은 개별 사용자, 열은 개별 아이템, 값은 평점으로 구성되며, **pivot_table()** 함수로 형태 변경

일반적으로 많은 아이템을 열로 가지는 다차원 행렬이며, 사용자가 평점을 매기는 경우가 많지 않아 희소 행렬 특성을 가짐

최근접 이웃 협업 필터링은 메모리 협업 필터링이라고도 하며, 사용자 기반과 아이템 기반으로 나눌 수 있음

- **사용자 기반 최근접 이웃 방식 : 특정 사용자와의 유사도가 높은 TOP-N 사용자가 좋아하는 아이템을 추천하는 방식**

(Customers like you also bought these items.)


- **아이템 기반 최근접 이웃 방식 : 사용자들의 평가 척도가 유사한 아이템을 추천하는 방식**

(Customers who bought this item also bought these items.)

일반적으로 사용자 기반보다는 아이템 기반 협업 필터링의 정확도가 높기 때문에 **대부분 아이템 기반의 알고리즘.** 주로 코사인 유사도 이용

## 1.4 잠재 요인 협업 필터링

- **잠재 요인 협업 필터링 방식 : 사용자-아이템 평점 행렬 속의 잠재 요인을 추출해 추천하는 방식**

대규모 다차원 행렬을 차원 감소 기법으로 분해하는 과정에서 잠재 요인을 추출하는 **행렬 분해** 기반의 잠재 요인 협업 필터링

사용자-아이템 행렬 데이터를 사용자-잠재 요인 행렬과 잠재 요인-아이템 행렬로 분해한 후,

분해된 두 행렬의 내적을 통해 새로운 예측 사용자-아이템 평점 행렬 데이터를 만들어 아이템에 대한 예측 평점을 생성

(잠재 요인이 정확히 어떤 것인지 알 수 없으나, 영화 평점 기반의 사용자-아이템 행렬에서 장르별 특성 선호도로 해석할 수 있음)

### 1.4.1 행렬 분해의 이해

- 행렬 분해 : 다차원의 매트릭스를 저차원 매트릭스로 분해하는 기법 ex) SVD, NMF

SVD는 NaN 값이 없는 행렬에만 적용할 수 있으므로, 주로 확률적 경사 하강법 방식을 이용해 SVD 수행

### 1.4.2 확률적 경사 하강법(SGD)을 이용한 행렬 분해

P와 Q 행렬로 계산된 **예측 R 행렬 값이 실제 R 행렬 값과 최소의 오류를 가질 수 있도록 비용 함수 최적화를 통해 P와 Q를 유추함**

1) P와 Q를 임의의 행렬로 설정

2) P와 Q.T 값을 곱해 예측 R 행렬을 계산하고, 실제 R 행렬과의 오류 값 계산

3) 오류 값을 최소화하도록 P와 Q 행렬을 적절한 값으로 업데이트

4) 2,3번 작업을 반복하면서 P와 Q 값을 업데이트하여 근사화

## 1.5 콘텐츠 기반 필터링 실습 - TMDB 5000 영화 데이터 세트

anaconda 창에서 kaggle datasets download -d tmdb/tmdb-movie-metadata 입력하여 데이터 다운로드

### 1.5.1 데이터 로딩 및 가공

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

movies = pd.read_csv('./data/tmdb-movie-metadata/tmdb_5000_movies.csv')
print(movies.shape)
movies.head(3)
```

```python
df = movies[['id','title','genres','vote_average','vote_count','popularity','keywords','overview']]
```

```python
pd.set_option('max_colwidth', 100)
df[['genres','keywords']][:1]
```

genres 칼럼에서 딕셔너리의 key인 'name'으로 개별 장르를 추출할 수 있으며, keywords 칼럼 역시 동일한 구조를 가짐

ast 모듈의 **literal_eval()** 함수를 이용하면 문자열을 list [dict1, dict2] 객체로 변환할 수 있음

```python
from ast import literal_eval

df['genres'] = df['genres'].apply(literal_eval)
df['keywords'] = df['keywords'].apply(literal_eval)
```

```python
df['genres'] = df['genres'].apply(lambda x: [y['name'] for y in x])
df['keywords'] = df['keywords'].apply(lambda x: [y['name'] for y in x])
df[['genres', 'keywords']][:1]
```

### 1.5.2 장르 콘텐츠 유사도 측정

1) 문자열로 변환된 genres 칼럼을 Count 기반으로 피처 벡터화 변환

2) 변환한 데이터 세트를 코사인 유사도를 통해 비교. 이를 위해 데이터 세트 레코드별로 타 레코드와 장르에서 유사도 값을 가지는 객체 생성

3) 장르 유사도가 높은 영화 중 평점이 높은 순으로 영화 추천

```python
from sklearn.feature_extraction.text import CountVectorizer

# CountVectorizer를 적용하기 위해 공백문자로 word 단위가 구분되는 문자열로 변환
df['genres_literal'] = df['genres'].apply(lambda x: (' ').join(x))
count_vec = CountVectorizer(min_df=0, ngram_range=(1,2))
genre_mat = count_vec.fit_transform(df['genres_literal'])
print(genre_mat.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

genre_sim = cosine_similarity(genre_mat, genre_mat)
print(genre_sim.shape)
print(genre_sim[:1])
```

**argsort()[:, ::-1]**을 이용해 높은 순으로 정렬된 비교 행 위치 인덱스 값을 가져오고 0번 값만 샘플로 추출

```python
genre_sim_sorted_idx = genre_sim.argsort()[:, ::-1]
print(genre_sim_sorted_idx[:1])
```

0번 레코드의 경우 3494번 레코드, 813번 레코드 순으로 유사도가 높음

### 1.5.3 장르 콘텐츠 필터링을 이용한 영화 추천

```python
# 장르 유사도에 따라 영화를 추천하는 함수 생성
def find_sim_movie(df, sorted_idx, title_name, top_n=10):
    title_movie = df[df['title'] == title_name]
    
    title_idx = title_movie.index.values
    similar_idx = sorted_idx[title_idx, :(top_n)]
    
    print(similar_idx)
    similar_idx = similar_idx.reshape(-1)     # 1차원 array로 변경
    
    return df.iloc[similar_idx]
```

```python
similar_movies = find_sim_movie(df, genre_sim_sorted_idx, 'The Godfather', 10)
similar_movies[['title', 'vote_average']]
```

영화 평점 정보인 vote_average 칼럼을 이용해 평점에 따라 필터링해서 최종 추천하는 방식으로 변경

그 전에 voto_average는 왜곡된 데이터를 가지고 있으므로, 이를 먼저 확인

```python
df[['title','vote_average','vote_count']].sort_values('vote_average', ascending=False)[:10]
```

생소한 영화들이 평가 횟수는 매우 작고 평점은 높은 왜곡된 데이터를 가지고 있음

IMDB에서 사용하는 평가 횟수에 대한 가중치가 부여된 평점 방식을 이용해 새롭게 평점 부여 (C: 전체 영화의 평균 평점, m: 최소 투표 횟수)

```python
C = df['vote_average'].mean()
m = df['vote_count'].quantile(0.6)     # 투표 횟수를 상위 60% 기준으로 정함
print('C: ', round(C, 3), 'm: ', round(m, 3))
```

```python
def weighted_vote_average(record):
    v = record['vote_count']
    R = record['vote_average']
    
    return ( (v/(v+m)) * R ) + ( (m/(m+v)) * C ) 

df['weighted_vote'] = df.apply(weighted_vote_average, axis=1)
```

```python
df[['title','vote_average','weighted_vote','vote_count']].sort_values('weighted_vote', ascending=False)[:10]
```

```python
# 새롭게 정의된 평점 기준에 따라 영화를 추천하는 함수 생성
def find_sim_movie(df, sorted_idx, title_name, top_n=10):
    title_movie = df[df['title'] == title_name]
    title_idx = title_movie.index.values
    
    # top_n의 2배에 해당하는 장르 유사성이 높은 인덱스 추출
    similar_idx = sorted_idx[title_idx, :(top_n * 2)]
    similar_idx = similar_idx.reshape(-1)
    # 기준 영화 인덱스는 제외
    similar_idx = similar_idx[similar_idx != title_idx]
    
    return df.iloc[similar_idx].sort_values('weighted_vote', ascending=False)[:top_n]

similar_movies = find_sim_movie(df, genre_sim_sorted_idx, 'The Godfather', 10)
similar_movies[['title', 'vote_average', 'weighted_vote']]
```

### 1.6.1 데이터 가공 및 변환

```python
import pandas as pd
import numpy as np

movies = pd.read_csv('./data/ml-latest-small/movies.csv')
ratings = pd.read_csv('./data/ml-latest-small/ratings.csv')
print(movies.shape)
print(ratings.shape)
movies.head(3)
```

```python
ratings.head(3)
```

**pivot_table()** 함수를 이용해 사용자를 로우로, 영화를 칼럼으로 구성한 데이터 세트로 변경

```python
ratings = ratings[['userId', 'movieId', 'rating']]
ratings_matrix = ratings.pivot_table('rating', index='userId', columns='movieId')
ratings_matrix.head(3)
```

```python
# 칼럼명을 영화명으로 변경하기 위해 movies와 조인
rating_movies = pd.merge(ratings, movies, on='movieId')

# columns='title'로 title 칼럼으로 피벗 수행
ratings_matrix = rating_movies.pivot_table('rating', index='userId', columns='title')

# 평점이 NaN인 값을 모두 0으로 변환
ratings_matrix = ratings_matrix.fillna(0)
ratings_matrix.head(3)
```
